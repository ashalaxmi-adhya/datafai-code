```python
#import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#basic statistics
iris.describe()
np.round(iris.describe(),2)
```

```python
np.round(iris.mean(),2)
```

```python
np.round(iris.median(),2)
```

```python
#Display non missing counts by each column
iris.count()
```

```python
#counting frequency for categorical variables
pd.crosstab(index=iris["species"],columns="Frequency")
```

```python
correlation = iris.corr()
correlation
```

```python
#Color table between variables for correlation
sns.heatmap(correlation,annot=True)
#cmap="RdYIGn",
plt.show()
sns.heatmap(correlation,cmap="Blues",annot=True)
#cmap="Blues",
plt.show()
```

```python
#Compute Percentile
np.round(iris.quantile((np.arange(0,1,0.05))),2) #from 0% to 100% in the gap of 5%
```

```python
#Pairplot
sns.pairplot(iris,hue="species",size=5)
#palette="RdYIGn",
plt.show()
```

```python
sns.pairplot(iris,hue='species',palette="copper",height=5,x_vars=["sepal_width","sepal_length"],y_vars=["petal_width","petal_length"])
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplots_adjust(hspace=0.2,wspace=0.2)
plt.subplot(2,2,1)
plt.title('Sepal Length')
plt.plot(iris.sepal_length,'r')
plt.subplot(2,2,2)
plt.title('Sepal Width')
plt.plot(iris.sepal_width,'b')
plt.subplot(2,2,3)
plt.title('Petal Length')
plt.plot(iris.petal_length,'g')
plt.subplot(2,2,4)
plt.title('Petal Width')
plt.plot(iris.petal_width,'y')
```

```python
#Histogram and Kernel Density Estimator
plt.figure(figsize=(15,10))
sns.distplot(iris.sepal_width)
plt.show()
```

```python
#Histogram and Kernel Density Estimator
plt.figure(figsize=(15,10))
sns.distplot(iris.sepal_width)
plt.show()
#Histogram and Kernel Density Estimator
plt.figure(figsize=(15,10))
sns.distplot(iris.sepal_length)
plt.show()
#Histogram and Kernel Density Estimator
plt.figure(figsize=(15,10))
sns.distplot(iris.petal_width)
plt.show()
#Histogram and Kernel Density Estimator
plt.figure(figsize=(15,10))
sns.distplot(iris.petal_length)
plt.show()
```

```python
#BoxPlot
plt.figure(figsize=(15,10))
sns.boxplot(data=iris,orient="v",palette="Set2")
plt.show()
```
