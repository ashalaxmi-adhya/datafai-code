```python
#import data
import numpy as np
import pandas as pd
df=pd.read_csv('beef_price_elasticity_data.csv')
np.round(df.describe(),2)
```

```python
#establish graphical relationship between demand and price of Beef
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.pairplot(df,kind='reg',height=10,x_vars=['Price'],y_vars=['Quantity'])
plt.show
```

```python
#Build log-log linear model
import statsmodels.api as sm
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings('ignore')
lm=smf.ols('np.log(Quantity)~np.log(Price)',data=df).fit()
```
