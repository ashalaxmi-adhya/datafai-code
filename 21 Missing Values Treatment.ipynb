```python
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
mydict={"Income":(50,25,15,10,np.NAN,np.NAN,20,10,np.NAN,999),"Age":(19,20,30,np.NAN,np.NAN,25,20,52,46,9999)}
print (mydict)
mydata=pd.DataFrame(mydict)
print(mydata)
```

```python
#Replace bogus values as 9999 with NaN
mydata=mydata.replace((9999,999),np.NAN)
print(mydata)
```

```python
#display values which are null with True
mydata.isnull()
```

```python
#Drop any rows with any columns missing values
new=mydata.dropna(how='any')
new
```

```python
#drop any rows where there are all columns having missing values
new=mydata.dropna(how='all')
new
```

```python
#Replace missing values with mean of respective columns
new =np.round(mydata.fillna(mydata.mean()),1)
new
```

```python
#replace missing values with median of respective columns
new =mydata.fillna(mydata.median())
new
```

```python
#Replace missing values with any numbers of your choice
new=mydata.fillna({'Age':30,'Income':15})
new
```
